```python
import os
DATASET_DIR = "D:/app/npdataset/yolo_dataset"

assert os.path.exists(DATASET_DIR), "‚ùå Dataset folder not found"
assert os.path.exists(f"{DATASET_DIR}/images/train"), "‚ùå images/train missing"
assert os.path.exists(f"{DATASET_DIR}/images/val"), "‚ùå images/val missing"
assert os.path.exists(f"{DATASET_DIR}/labels/train"), "‚ùå labels/train missing"
assert os.path.exists(f"{DATASET_DIR}/labels/val"), "‚ùå labels/val missing"
assert os.path.exists(f"{DATASET_DIR}/data.yaml"), "‚ùå data.yaml missing"

print("‚úÖ Dataset structure verified")
```

```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8n.pt")
print("‚úÖ YOLOv8 model loaded")

results = model.train(
    data=f"{DATASET_DIR}/data.yaml",
    epochs=50,        # increase later (100+)
    imgsz=640,
    batch=16,         # reduce if RAM issue
    workers=4,
    device="cpu",     # use "0" if you have NVIDIA GPU
    project="runs",
    name="license_plate_yolov8",
    exist_ok=True
)

print("üéâ Training completed")
```

```python
model.export(format="onnx")
print("‚úÖ Model exported to ONNX format")
```

```python
model = YOLO("D:\\app\\npdataset\\runs\\detect\\runs\\license_plate_yolov8\\weights\\best.pt")

results = model.predict(
    source="D:/app/npdataset/yolo_dataset/images/val",
    conf=0.25,
    save=True
)

print("‚úÖ Predictions complete")
```
